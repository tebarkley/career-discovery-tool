This notebook takes in a pickle file of raw inputs from the Smartrecruiters API and outputs a dataframe indexed by the job ID containing the following columns: text, sentences, and words. The purpose is to put data into a standardized format for downstream analysis.

```python
import nltk
import re
from os import path, pardir
import pandas as pd
from pandas import DataFrame, Series
import numpy as np
from bs4 import BeautifulSoup
```

```python
def visible_text(element):
    if element.parent.name in ['style', 'script', '[document]', 'head', 'title']:
        return ''
    #get rid of tags
    result = re.sub('<br>|<br \/>|<p\/?>|<\/?li>', '\n', str(element.encode('ascii', 'ignore')))
    result = re.sub('<.*?>', ' ', result, flags=re.DOTALL)
    #add single quotes
    result = re.sub('&rsquo;', "'", result)
    #get rid of extra spaces, html spaces, and bullets
    result = re.sub(' {2,}|&nbsp;|&bull;', ' ', result)
    #replace multiple newlines with single newline
    result = re.sub('\s{2,}|\t', '\n', result)
    #add in double quotes
    result = re.sub('&ldquo;', '"', result)
    result = re.sub('&rdquo;', '"', result)
    result = re.sub('&rdquo;', '"', result)
    return result

#function for cleaning up HTML into text
def rawToText(description):
    soup = BeautifulSoup(description)
    texts = soup.findAll(text=True)
    visible_elements = [visible_text(elem) for elem in texts]
    text = ''.join(visible_elements)
    return text

#function for splitting sentences on new lines
def splitNewLine(sents):
    final_sents = []
    for sent in sents:
        sub_sents = sent.split('\n')
        for sub in sub_sents:
            if sub != '':
                final_sents.append(sub)
    return final_sents

#function for splitting sentences on semi-colon
def splitSemicolon(sents):
    final_sents = []
    for sent in sents:
        sub_sents = sent.split(';')
        for sub in sub_sents:
            if sub != '':
                final_sents.append(sub)
    return final_sents
```

```python
def tokenizeCB(path_to_pkl_df, path_to_output_df):
    data = pd.read_pickle(path_to_pkl_df)
    #create simplified dataframe with just post ID and job description text
    post = data[['id', 'jobAd.sections.jobDescription.text']]
    post.set_index('id', inplace=True)
    post.columns = ['raw_text']
    #drop rows containing NaN
    post.dropna(inplace=True)
    #Add a column with better HTML
#     post['html'] = post.raw_text.map(lambda x: BeautifulSoup(x))

    post['html'] = post.raw_text.map(lambda x: BeautifulSoup(x).getText())
    #Add a column with just text:  http://stackoverflow.com/questions/10524387/beautifulsoup-get-text-does-not-strip-all-tags-and-javascript
    post['text'] = post.raw_text.apply(rawToText)
    #Add a column containing a list of sentences
    post['sents'] = post.text.map(lambda x: nltk.sent_tokenize(x))
    #Further split sentences on new lines
    post['sents'] = post['sents'].map(splitNewLine)
    post['sents'] = post['sents'].map(splitSemicolon)
    #Add a column with list (sentences) of list (words)
    post['words'] = post.sents.map(lambda x: [nltk.word_tokenize(sent) for sent in x])
    final = post[['text','sents','words']]
    final.to_pickle(path_to_output_df)
    return post
```

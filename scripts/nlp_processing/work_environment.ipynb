# Extract text for Stanford pattern learner

Create a text file to feed into the Stanford tagger. We will start by pulling out 100 random job descriptions, hand identifying some work environment phrases, adjusting the parameters, and seeing how well it does. Feeding the pattern learner the raw text did not do well. We'll try joining our sentence tokenization with periods. 

```python
workenv = data.reindex(np.random.permutation(data.index))[:100]
workenv.head()
```

```python
workenv_sents = workenv.sents.map(lambda x: '. '.join(x)).tolist()
workenv_string = '.\n'.join(workenv_sents)
```

```python
with open('workenvs.txt','wb') as f:
    f.write(workenv_string)
f.close()
```

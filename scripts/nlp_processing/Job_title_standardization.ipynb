# Normalizing Job Posting Titles 

This notebook extracts normalized job titles across O*NET, CareerBuilder, Craigslist, and Smart Recruiters data sources.

```python
import nltk
import re
from os import path, pardir
import pandas as pd
from pandas import DataFrame, Series
import numpy as np
from collections import defaultdict
```

###Preparation: dedupe job description metadata

```python
all_job_descriptions_df = pd.read_pickle('../../data/all_jobs_combined.pkl')
```

####Get list of deduped job IDs for each data source

```python
deduped_cl_ids = list(all_job_descriptions_df[all_job_descriptions_df.dataset == 'all_craigslist_tokenized_dedup.pkl'].index)
deduped_sr_ids = list(all_job_descriptions_df[all_job_descriptions_df.dataset == 'smartrecruiter_all_text.pkl'].index)
deduped_cb_ids = list(all_job_descriptions_df[all_job_descriptions_df.dataset == 'cb_data_all_text.pkl'].index)
```

####Filter each metadata dataframe on respective IDs to remove duplicates

```python
all_craigslist_df = pd.read_pickle('../../data/all_craigslist_df.pkl')
craigslist_filtered_df = all_craigslist_df[np.in1d(all_craigslist_df.url, deduped_cl_ids)]

# craigslist_filtered_df.head()
```

```python
all_sr_df = pd.read_pickle('../../data/smartrecruiter_all_dedup.pkl')
sr_filtered_df = all_sr_df[np.in1d(all_sr_df.index, deduped_sr_ids)]

# sr_filtered_df.head()
```

```python
all_cb_df = pd.read_pickle('../../data/cb_data_all_dedup.pkl')
cb_filtered_df = all_cb_df[np.in1d(all_cb_df.index, deduped_cb_ids)]

# cb_filtered_df.head()
```

##Job Title Extraction

###Get Normalized list of CareerBuilder Job Titles to cross-ref with O*NET codes

```python
def normalize_raw_title(raw_job_title):
    normed_title = re.sub(r'''[0-9]*''', '', raw_job_title).lower().strip()
    normed_title = re.sub(r'''[\.,;"'$?!():\-_`/+#*]''', ' ', normed_title)
    normed_title = re.sub(r'''\s[i]+''', ' ', normed_title)
    normed_title = re.sub(r'''\s[i]+$''', ' ', normed_title)
    normed_title = re.sub(r'''[\s]{2,}''', ' ', normed_title)
    return normed_title

# dictionary of {onet_code: onet_friendly_cb_title, cb_title, normalized_cb_title}  
# NOTE: only 435/10,025 CB jobs are tagged with ONET codes
cb_job_map = Series(zip(cb_filtered_df.ONetFriendlyTitle.values, cb_filtered_df.JobTitle.values), 
                    index=cb_filtered_df.OnetCode).to_dict()

for key,value in cb_job_map.items():
    cb_job_map[key] = list(value) # cast tuple to list so that it is mutable
    cb_job_map[key].append(normalize_raw_title(value[1])) # append normalized title to value list

# full list of CB normalized job titles
cb_normed_job_titles = cb_filtered_df.JobTitle.apply(normalize_raw_title)
cb_normed_job_titles = list(cb_normed_job_titles.values) # convert from Series to list
```

### O*NET Data

```python
onet_df = pd.read_csv('../../data/onet_data_normalized_2.csv')
```

```python
def normlize_onet_title(raw_onet_title):
    
    # replace underbars with spaces
    normed_title = raw_onet_title.replace('_', ' ')
    
    # common position words that start with 's' in the middle of a title
    title_snippets = ['administrators', 'examiners', 'adjusters', 'appraisers', 'agents', 
                      'postmasters', 'investigators', 'analysts', 'scientists', 'biochemists'
                      'technicians', 'directors', 'teachers', 'cooks', 'artists', 'assistants']

    # normalize spaces, remove punctuation, numbers
    normed_title = re.sub(r'''[0-9]*''', '', normed_title).lower().strip()
    normed_title = re.sub(r'''[\.,;"'$?!():\-_`/+#*]''', ' ', normed_title)
    normed_title = re.sub(r'''[\s]{2,}''', ' ', normed_title)

    if normed_title.split()[-1].endswith('s'):
        normed_title = normed_title[:-1] # remove trailing 's'
    
    for item in title_snippets:
        if item in normed_title:
            normed_title = normed_title.replace(item, item[:-1]) # remove trailing 's'
    
    #manually fix any other idiosyncrasies here
    if title == 'midwive':
        normed_title = 'midwife'
    
    return normed_title
```

#### Create O*NET map from codes to titles

```python
onet_titles = Series(onet_df.title.values,index=onet_df.element_name).to_dict()

for key,value in onet_titles.items():
    onet_titles[key] = [value]
    onet_titles[key].append(normlize_onet_title(value))
    
# dictionary of {onet_code: onet_dataset_title, normed_onet_title, [cb_titles]}    
onet_titles
```

```python
def inferred_onet_code(x):
    onet_codes = []
    normed_target_text = normalize_raw_title(x)
    for key,values in onet_titles.items():
        if values[1] in normed_target_text:
            onet_codes.append(key)
            
    for key,values in cb_job_map.items():
        if values[2] in normed_target_text:
            onet_codes.append(key)
            
    return list(set(onet_codes))

def inferred_job_title(x):
    job_titles = []
    normed_target_text = normalize_raw_title(x)
    for key,values in onet_titles.items():
        if values[1] in normed_target_text:
            job_titles.append(values[1].lower())
    
    for title in cb_normed_job_titles:
        if title in normed_target_text:
            job_titles.append(title)
    
    return list(set(job_titles))


# New columns with inferred standardized titles and O*NET codes 
craigslist_filtered_df['onet_codes'] = craigslist_filtered_df.title.apply(inferred_onet_code)
craigslist_filtered_df['standardized_titles'] = craigslist_filtered_df.title.apply(inferred_job_title)
craigslist_filtered_df['possible_titles_from_desc'] = craigslist_filtered_df.description.apply(inferred_job_title)

sr_filtered_df['onet_codes'] = sr_filtered_df.name.apply(inferred_onet_code)
sr_filtered_df['standardized_titles'] = sr_filtered_df.name.apply(inferred_job_title)
sr_filtered_df['possible_titles_from_desc'] = sr_filtered_df.text.apply(inferred_job_title)

cb_filtered_df['onet_codes'] = cb_filtered_df.JobTitle.apply(inferred_onet_code)
cb_filtered_df['standardized_titles'] = cb_filtered_df.JobTitle.apply(inferred_job_title)
cb_filtered_df['possible_titles_from_desc'] = cb_filtered_df.text.apply(inferred_job_title)
```

# Getting Data from JobSearch CB API

```python
import pandas as pd
import numpy as np
import requests
import xmltodict
import json
from pandas.io.json import json_normalize
import settings
from os import pardir, path
from __future__ import division
```

```python
def get_jobsearch_data(page_num):
    parameters = {'DeveloperKey': settings.CB_API_KEY_TIFF, 
                  'Location': 'San Francisco, CA',
                  'PostedWithin': 7,
                  'Radius': 50,
                  'PerPage': 100,
                  'PageNumber': page_num}

    r = requests.get('http://api.careerbuilder.com/v2/jobsearch/', params=parameters)


    o = xmltodict.parse(r.content)
    j = json.dumps(o)
    raw_df = pd.read_json(j)
    new_df = json_normalize(raw_df.T.Results, 'JobSearchResult') # http://pandas.pydata.org/pandas-docs/stable/io.html

    return new_df
```

```python
for page in range(1,33): # expected 11,033 jobs in query, so requested 111 pages
    if page == 1:
        master_df = get_jobsearch_data(page)
    else:
        master_df = pd.concat([master_df, get_jobsearch_data(page)], ignore_index=True)
```

```python
import pandas as pd
import numpy as np
import requests
import json
import settings
import sys
import csv
from os import pardir, path
from pandas.io.json import json_normalize
from __future__ import division
```

```python
postings = pd.read_csv(path.join(pardir, pardir, 'data', 'bay_area_company_job_ids.csv'))
postings.tail(10)
```

```python
#create list of company, posting id tuples
company_id = zip(postings.company.tolist(), postings.id.tolist())
```

```python
#define function for calling job postings api
def getJobPost(company_name, posting_id):
    r = requests.get('https://api.smartrecruiters.com/v1/companies/' + company_name + '/postings/' + str(posting_id))
    j = json.loads(r.content)
    df = json_normalize(data = j)
    return df
```

```python
all_postings = pd.DataFrame()
for company, job_id in company_id:
    try:
        df = getJobPost(company, job_id)
    except ValueError:
        continue
    all_postings = pd.concat([all_postings, df], ignore_index=True)
    #write out every 500 postings to a pkl file just in case
    num_jobs = len(all_postings)
    if num_jobs%500 == 0:
        all_postings.to_pickle(path.join(pardir, pardir, 'data', 'job_postings_' + str(num_jobs)+'.pkl'))    
```

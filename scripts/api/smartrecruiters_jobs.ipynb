```python
import pandas as pd
import numpy as np
import requests
import json
import settings
import sys
import csv
from os import pardir, path
from __future__ import division
```

```python
companies = pd.read_csv('companies_with_jobs_in_ca.csv', header=None)[0].tolist()
```

```python
##remove companies from which we have already obtained job posting ids
obtained = pd.read_csv(path.join(pardir, pardir, 'data','company_job_ids.csv'), header=None)
companies_obtained = set(obtained[0].values)
companies_obtained
companies = [company for company in companies if company not in companies_obtained]
```

```python
def get_jobsearch_data(num_jobs, company_name):
    parameters = {'offset': num_jobs,
                'region': 'CA'}
    r = requests.get('https://api.smartrecruiters.com/v1/companies/' + company_name + '/postings', params=parameters)
    j = json.loads(r.content)
    job_ids = [obj['id'] for obj in j['content']]
    job_cities = [obj['location']['city'] for obj in j['content']]
    return job_ids, job_cities
```

Obtain company job posting ids and cities

```python
for company in companies:
    company_job_ids_all = []
    company_job_cities_all = []
    while True:
        num_company_jobs = len(company_job_ids_all)
        job_ids_run, job_cities_run = get_jobsearch_data(num_company_jobs, company)
        company_job_ids_all.extend(job_ids_run)
        company_job_cities_all.extend(job_cities_run)
        #if the job call returns fewer than 100 job ids, move on to the next company
        if len(job_ids_run) < 100:
            #write out job ids and cities to a text file
            company_and_ids = zip([company]*len(company_job_ids_all), company_job_ids_all, company_job_cities_all)
            with open(path.join(pardir, pardir, 'data','company_job_ids.csv'), "a") as the_file:
                writer = csv.writer(the_file, delimiter=',', lineterminator='\n')
                for row in company_and_ids:
                    try:
                        writer.writerow(row)
                    except UnicodeEncodeError:
                        continue
            break
```
